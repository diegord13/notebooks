MISSING DATA

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("miss").getOrCreate()
```

```python
df = spark.read.csv("ContainsNull.csv",header=True,inferSchema=True)
```

```python
df.show()
```

```python
df.na.drop(thresh=2).show()
```

```python
df.na.drop(how="all").show()
```

```python
df.na.drop(subset=["Sales"]).show()
```

```python
df.na.fill("FILL VALUE",subset=["Name"]).show()
#solo llena valores en la columna del tipo string si se ingresa strings en los valores nulos
```

```python
df.na.fill(0).show()
#Solo llena valores numericos en la columna con este tipo de valor y con valores nulos
```

```python
from pyspark.sql.functions import mean
```

```python
mean_val = df.select(mean(df["Sales"])).collect()
```

```python
mean_sales =mean_val[0][0]
```

```python
df.na.fill(mean_sales,["Sales"]).show()
```

```python
df.na.fill(df.select(mean(df["Sales"])).collect()[0][0],["Sales"]).show()
#la misma linea anterior en una sola fila rellenando en valores numericos con valor nulo el valor del promedio hallado en mean_sales
```

DATES AND TIMESTAMPS

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("dates").getOrCreate()
```

```python
df = spark.read.csv("appl_stock.csv",header=True,inferSchema=True)
```

```python
df.head(1)
```

```python
df.show()
```

```python
df.select(["Date","Open"]).show()
```

```python
from pyspark.sql.functions import (dayofmonth,hour,
                                   dayofyear,month,
                                   year,weekofyear,
                                   format_number,date_format)
```

```python
df.select(dayofmonth(df["Date"])).show()
```

```python
newdf = df.withColumn("Year",year(df["Date"]))
```

```python
result = newdf.groupby("Year").mean().select(["Year","avg(Close)"])
```

```python
new = result.withColumnRenamed("avg(Close)","Average Closing Price")
```

```python
new.select(["Year",format_number("Average Closing Price",2).alias("Avg Close")]).show()
```
